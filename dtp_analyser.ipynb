# Табличка крафтится

```python
%matplotlib inline

import numpy as np
import pandas as pd
from matplotlib import pyplot, rcParams
import time
```

```python
dataset = pd.read_table("data_dtp/cards.tsv")
```

```python
df = dataset[dataset.city == 'г Москва']
```

```python
df
```

```python
df.street = df.street.fillna(0)
df.house = df.house.fillna(0)
df.road = df.road.fillna(0)
df.km = df.km.fillna(0)
df.m = df.m.fillna(0)
df
```

```python
year = []
month = []
day = []
hours = []
minutes = []
for i in range(15785):
    info = df.iloc[i].timestamp
    year.append(int(info[:4]))
    month.append(int(info[5:7]))
    day.append(int(info[8:10]))
    hours.append(int(info[11:13]))
    minutes.append(int(info[14:16]))

df['year'] = np.array(year)
df['month'] = np.array(month)
df['day'] = np.array(day)
df['hours'] = np.array(hours)
df['minutes'] = np.array(minutes)
```

```python
address = []
for i in range(15785):
    address.append(str(df.iloc[i].street) + ' ' + str(df.iloc[i].house))
df['address'] = np.array(address)
df
```

```python
result = []
for i in range(15785):
    info = df.iloc[i]
    if info.fatalities:
        result.append(2)
    elif info.victims:
        result.append(1)
    else:
        result.append(0)
df['result'] = np.array(result)
```

```python
daytime = []
for i in range(15785):
    daytime.append(int(df.iloc[i].hours >= 19 or df.iloc[i].hours <= 5))
df['daytime'] = np.array(daytime)
```

```python
weekday = []
for i in range(15785):
    weekday.append(time.strptime(df.iloc[i].timestamp, "%Y-%m-%d %H:%M:%S")[6])
df['weekday'] = np.array(weekday)
```

```python
df
```

# Визуализация

```python
tclass = df.groupby(['hours', 'result']).size().unstack()
```

```python
tclass
```

```python
pyplot.figure(figsize=(12, 8))
pyplot.bar(list(range(24)), tclass[1] + tclass[2])
#pyplot.bar(list(range(24)), tclass[2], bottom=tclass[1], label='Fatal')
pyplot.xticks(np.array(range(24)) + 0.5, list(range(24)), rotation='horizontal')
pyplot.ylabel("Number")
```

```python
pyplot.figure(figsize=(12, 8))
pyplot.bar(list(range(24)), 100 * tclass[2] / (tclass[1] + tclass[2]), color = '#e0c92f')
pyplot.xticks(np.array(range(24)) + 0.5, list(range(24)), rotation='horizontal')
pyplot.ylabel("Percentage of fatalities (%)")
```
